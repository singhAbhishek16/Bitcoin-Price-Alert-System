```python
import requests
import json
import time

### Configuration file for Bitcoin Price Alert System

telegram_chat_id = "@bitcoin_alert_channel"
telegram_bot_id = "bot1389359250:AAFOP_PfH2gf8FBJcfQ7wEPIl01SzjNJ6Jw"
url = "https://min-api.cryptocompare.com/data/price?fsym=BTC&tsyms=INR"

### Enter predicted Bitcoin Price Below
sp_prediction = 820000


def get_bitcoin_price():
    
    response = requests.request("GET",url)
    response = json.loads(response.text)
    current_price = response["INR"]
    return current_price



def send_telegram_message(message):
    
    url = "https://api.telegram.org/" + telegram_bot_id + "/sendMessage"
    data = {
        "chat_id": telegram_chat_id,
        "text": message
    }
    try:
        response = requests.request(
            "POST",
            url,
        params = data
        )
        print("This is the telegram URL")
        print(url)
        print(response.text)
        telegram_data = json.loads(response.text)
        return telegram_data["okay"]
    
    except Exception as e:
        
        print("Some error occured while sending telegram alert")
        print(e)
        return False

    
while True:
    
    print("Obtaining Bitcoin price...")
    time.sleep(1)
    curr_price = get_bitcoin_price()
    print("It's " + str(curr_price) + "INR")
    time.sleep(1)
    
    try:
        if curr_price >= sp_prediction:
            print("Bitcoin price has exceeded your set selling price limit by ",curr_price-sp_prediction," INR")
            time.sleep(1)
            print("Alert ON")
            message = "Alert! Bitcoin value has exceeded your set selling price by " + str(curr_price-sp_prediction) + " INR." + "The current value is " + str(curr_price) + "INR"
            telegram_status = send_telegram_message(message)
            print("This is the telegram_status: ", telegram_status)
            time.sleep(1)
            print("Switching OFF alert...")
            
        else:
            print("No alert generated. Bitcoin price is less than set selling price.")
            
    except Exception as e:
        
        print("Error occured. Below are the details")
        print(e)
        
    time.sleep(1)
    print("Will check the price again in 30 seconds. You will be updated via telegram notification.")
    time.sleep(30)
```
